```python
from pyts.image import GramianAngularField
from pyts.datasets import load_gunpoint
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# darkgrid
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
```

```python
#load TS data
X, _, _, _ = load_gunpoint(return_X_y=True)
```

```python
# Apply transformation
transformer = GramianAngularField(image_size=24)
X_img = transformer.fit_transform(X)
```

```python
# Plot results
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
axes[0].plot(X[0])
axes[0].set_title('Original time series data')
axes[1].imshow(X_img[0], cmap='rainbow', aspect='auto')
axes[1].grid(False)
axes[1].set_title('Transformed time series data')
axes[1].axis('off')
# give a name to the figure at the bottom
plt.suptitle('Comparison of Original and Transformed Time Series Data', fontsize=14, y=0.05)
```
